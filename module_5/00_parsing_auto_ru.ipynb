```python
# Импортир библиотек
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from time import sleep, time
import pandas as pd
from random import randint
from datetime import datetime
```

## Функции для сбора данных

```python
def data_collecting(auto_all, auto_collection):
    
    """ Сбор данных с полученной страницы списка объявлений и запись их в список, 
    который определенн глобально и передается вторым аргументом.
    Функция ничего не возвращает, а добавляет сформированный из признаков словарь в список
    """
    
    
    for elem in auto_all:

        try: car_url = elem.find('a', class_="Link ListingItemTitle__link" ).get('href')
        except: car_url = None

        try: name = elem.find('a', class_="Link ListingItemTitle__link" ).text
        except: name = None

        param = elem.find_all('div', class_="ListingItemTechSummaryDesktop__cell")

        try: engineDisplacement = param[0].text.split(sep='/')[0].replace('\u2009', '')
        except: engineDisplacement = None                    

        try: enginePower = param[0].text.split(sep='/')[1].replace('\u2009', '').replace('\xa0', '')
        except:enginePower= None                    

        try: fuelType = param[0].text.split(sep='/')[2].replace('\u2009', '')
        except: fuelType = None                    

        try: vehicleTransmission = param[1].text
        except: vehicleTransmission = None                    

        try: bodyType = param[2].text
        except: bodyType = None                    

        try: Привод = param[3].text
        except: Привод = None                    

        try: color = param[4].text
        except: color = None                    

        try: productionDate = elem.find('div', class_='ListingItem__year').text
        except: productionDate = None                    

        try: mileage = int(elem.find('div', class_='ListingItem__kmAge').text.replace('\xa0', '').replace('км', ''))
        except: mileage = None                    

        try: price = int(elem.find('div', class_='ListingItemPrice__content').text.replace('\xa0', '').replace('₽', ''))
        except: price = None  

        auto_collection.append(
        {
            'car_url': car_url,
            'name': name,
            'engineDisplacement': engineDisplacement,
            'enginePower': enginePower,
            'fuelType': fuelType,
            'vehicleTransmission': vehicleTransmission,
            'bodyType': bodyType,
            'Привод': Привод,
            'color': color,
            'productionDate': productionDate,
            'mileage': mileage,
            'price': price
        })
       
    
   
```

```python
def data_from_ad_auto(soup, url, auto_collection_ad):
    
    """ Сбор данных с полученной страницы объявления и добавление их в список в виде словаря. 
    Список создан в глобальной области, который передается аргументом. 
    Аргумент url передет ссылку для признака ['car_url']
    Функция ничего не возращает.
    """

    param = soup.find_all('div', class_="InfoPopup InfoPopup_theme_plain InfoPopup_withChildren BreadcrumbsPopup")
    
    try: bodyType = soup.find('li', class_="CardInfoRow_bodytype").find('a').text
    except: bodyType = None

    try: brand = soup.find('h1').text.split(maxsplit=2)[0]
    except: brand = None

    try: car_url = url
    except: car_url = None

    try: color = soup.find('li', class_="CardInfoRow_color").find('a').text
    except: color = None

    try: description = soup.find('div', class_="CardDescription__textInner CardDescription__textInner_cut").text
    except: description = None

    try: engineDisplacement = soup.find('li', class_="CardInfoRow_engine").find('div').text.split('/')[0]
    except: engineDisplacement = None

    try: enginePower = soup.find('li', class_="CardInfoRow_engine").find('div').text.split('/')[1]
    except: enginePower = None

    try: fuelType = soup.find('li', class_="CardInfoRow_engine").find('a').text
    except: fuelType = None

    try: mileage = int(soup.find('li', class_="CardInfoRow_kmAge").text.replace('Пробег', '').replace('км', '').replace('\xa0', ''))
    except: mileage = None

    try: modelDate = param[2].text
    except: modelDate = None

    try: model_name = soup.find('h1').text.split(maxsplit=2)[1]
    except: model_name = None

    try: name = param[4].text
    except: name = None

    try: parsing_unixtime = round(time())
    except: parsing_unixtime = None

    try: productionDate = int(soup.find('li', class_='CardInfoRow_year').find('a').text)
    except: productionDate = None

    try: sell_id = int(soup.find('div', class_="CardHead__infoItem CardHead__id").text.replace('№ ', ''))
    except: sell_id = None

    try: vehicleTransmission = soup.find('li', class_='CardInfoRow_transmission').text.replace('Коробка', '')
    except: vehicleTransmission = None

    try: Владельцы = soup.find('li', class_='CardInfoRow_ownersCount').text.replace('Владельцы', '')
    except: Владельцы = None

    try: ПТС = soup.find('li', class_='CardInfoRow_pts').text.replace('ПТС', '')                                                           
    except: ПТС = None

    try: Привод = soup.find('li', class_='CardInfoRow_drive').text.replace('Привод', '')                                                      
    except: Привод = None

    try: Руль = soup.find('li', class_='CardInfoRow_wheel').text.replace('Руль', '')
    except: Руль = None
        
    try: Состояние = soup.find('li', class_='CardInfoRow_state').text.replace('Состояние', '')
    except: Состояние = None    

    try: price = int(soup.find('span', class_='OfferPriceCaption__price').text.replace('\xa0', '').replace('₽', ''))
    except: price = None

    auto_collection_ad.append(
        {
            'bodyType': bodyType,
            'brand': brand,
            'car_url': car_url,
            'color': color,
            'description': description,
            'engineDisplacement': engineDisplacement,
            'enginePower': enginePower,
            'fuelType': fuelType,
            'mileage': mileage,
            'modelDate': modelDate,
            'model_name': model_name,
            'name': name,
            'parsing_unixtime': parsing_unixtime,
            'productionDate': productionDate,
            'sell_id': sell_id,
            'vehicleTransmission': vehicleTransmission,
            'Владельцы': Владельцы,
            'ПТС': ПТС,
            'Привод': Привод,
            'Руль': Руль,
            'Состояние': Состояние,
            'price': price
        }
    )
    
    
```

```python
# Создаем список для сбора информации об объявлениях
auto_collection = []

# Запускаем chromedriver
browser = Chrome('C:/Users/User/Desktop/chromedriver') 

# Указываем ссылку с какой страницы начать парсить (выше приводились примеры)
url = 'https://auto.ru/cars/honda/accord/used/'
browser.get(url)
sleep(7) # задержка, чтоб прогрузился браузер

# В range подставляем количество страниц, которые нам нужно спарсить (смотрим на сайте).
# в зависимости от количества объявлений по этой марке автомобиля
# но не более 99, так как потом о гонит дубли
for page in range(20):
    print(f'Страница {page}')
    soup = BeautifulSoup(browser.page_source, 'lxml')

    # Выбираем объявления о продаже авто
    auto_all = soup.find_all('div', class_='ListingItem__main')

    # собираем признаки с внешних страниц, не погружаясь внутрь объявления 
    data_collecting(auto_all, auto_collection) 

    print(f'Элементов в списке: {len(auto_collection)}')
    
    # Переход на следующую страницу
    browser.find_element_by_class_name('ListingPagination__next').click()
    sleep(randint(7,11))
```

##  Код для парсинга страниц объявлений

Формируем список ссылок

```python
# Считываем файл со ссылками
df_for_link = pd.read_csv('for_processing.csv')
df_for_link.head(3)   
```

```python
# Достаем список ссылок, которые нам надо спарсить
links = list(df_for_link['car_url'])
```

Парсер.

```python
# Создаем список для сбора информации об объявлениях
auto_collection_ad = []

error_list = [] # список ошибок из словаря {count: url}
count = -1 # счетчик операций
```

```python
# Парсим страницы:

# Запускаем chromedriver
browser = Chrome('C:/Users/User/Desktop/chromedriver') 

# Запускаем цикл захода на каждую страницу объявления
for url in links:
    count +=1
    try: 
        browser.get(url)
    except Exception as e:
        print(e)
        print(f'На итерации {count} не получили даные по ссылке {url}')
        error_list.append({count: url})
        continue
     
        
    sleep(randint(7,11)) # задержка, чтоб прогрузился браузер
    
    soup = BeautifulSoup(browser.page_source, 'lxml')
    
    # собираем данные
    data_from_ad_auto(soup, url, auto_collection_ad)
    
    print(f'Элементов в списке: {len(auto_collection_ad)}')


```

```python
# Формируем датасет
df_auto_base = pd.DataFrame(auto_collection_ad)
df_auto_base.head(3)
```
